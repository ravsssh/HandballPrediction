# **PRE PROCESSING DATA**

```python
def extract_year_from_date(dataframe, date_column):
    dataframe[date_column] = pd.to_datetime(dataframe[date_column])
    dataframe[date_column] = dataframe[date_column].dt.year
    return dataframe

def convert_to_int(dataframe, column_name):
    dataframe[column_name] = dataframe[column_name].astype(int)
    return dataframe

def extract_hour_from_last_visit(dataframe, time_column):
    dataframe[time_column] = pd.to_datetime(dataframe[time_column])
    dataframe[time_column] = dataframe[time_column].dt.hour
    return dataframe
```

Kolom diatas adalah pembuatan fungsi yang nanti nya akan tinggal dipanggil untuk tujuan preprocessing data.


*   pada fungsi yang pertama bertujuan untuk mengubah kolom menjadi datetime dan mengambil hanya tahunnya
*   Pada fungsi yang kedua bertujuan untuk mengubah suatu kolom menjadi dattype integer
*   Pada fungsi yang ketiga bertujuan untuk mengubah kolom menjadi datetime dan mengambil hanya jam nya






```python
# Memanggil fungsi untuk mengubah kolom 'joining_date' menjadi tipe datetime dan mengekstrak tahunnya
Data = extract_year_from_date(Data, 'joining_date')


Data = convert_to_int(Data, 'avg_frequency_login_days')

Data = extract_hour_from_last_visit(Data, 'last_visit_time')
```

```python
def normal(df,col):
    upper = df[col].mean() + (3 * df[col].std())
    lower = df[col].mean() - (3 * df[col].std())

    outlier = df[(df[col] <= lower) | (df[col] >= upper)]

    print('percentage: ', len(outlier)/df.shape[0] * 100, '%')

def skew(df, col, fold):
    Q1 = df[col].quantile(.25)
    Q3 = df[col].quantile(.75)

    IQR = Q3 - Q1

    upper = Q3 + (fold*IQR)
    lower = Q1 - (fold*IQR)

    outlier = df[(df[col] <= lower) | (df[col] >= upper)]

    print('percentage: ', len(outlier)/df.shape[0] * 100, '%')
    print('upper: ', upper)
    print('lower: ', lower)

#Function to plot distribution
def plot_dist(df, col):
    fig, axes = plt.subplots(ncols = 2, figsize = (15, 5))

    # histogram
    sns.histplot(df[col],ax = axes[0], bins=30)
    axes[0].set_title(f"Histogram '{col}'")
    axes[0].axvline(df[col].mean(), color = 'red', linestyle = 'dashed', label = 'mean')
    axes[0].axvline(df[col].median(), color = 'green', linestyle = 'dashed', label = 'median')
    axes[0].legend()

    # boxplot
    sns.boxplot(y=df[col], ax =  axes[1])
    axes[1].set_title(f"Boxplot '{col}'")

    plt.show()

    # skewness
    print(df[col].name + ' Kurtosis: ' + str(df[col].kurt()))
    print(df[col].name + ' Skewness: ' + str(df[col].skew()))
    if -0.5 <= df[col].skew() <= 0.5:
        print("Columns '{}' normal distribution".format(col))
        normal(df, col)
    elif df[col].skew() > 0.5:
        if df[col].skew() > 1:
            print("Columns '{}' extreme right skewed".format(col))
            skew(df, col, 3)
        else:
            print("Columns '{}' right skewed".format(col))
            skew(df, col, 1.5)
    elif df[col].skew() < -0.5:
        if df[col].skew() < 1:
            print("Columns '{}' extreme left skewed".format(col))
            skew(df, col, 3)
        else:
            print("Columns '{}' left skewed".format(col))
            skew(df, col, 1.5)
```

```python
# Membuat cat_pipeline dengan label encoding
cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Mengisi nilai yang hilang (jika ada)
    ('ordinalEncoder', OrdinalEncoder())  # Label encoding pada fitur kategorikal
])
cat_tr = cat_pipeline.fit_transform(X_train[[
 'gender',
 'region_category',
 'membership_category',
 'joined_through_referral',
 'preferred_offer_types',
 'medium_of_operation',
 'internet_option',
 'used_special_discount',
 'offer_application_preference',
 'past_complaint',
 'complaint_status',
 'feedback']])
```

melakukan encoding terhadap data yang kategorikal

```python
num_pipeline = Pipeline([
        ('outlier', Winsorizer(capping_method='iqr', tail='both', fold=3)),
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', MinMaxScaler())
    ])

num_tr = num_pipeline.fit_transform(X_train[['points_in_wallet', 'avg_time_spent', 'avg_transaction_value']])
num_tr

```

melakukan handling outlier serta scaling terhadap data yang memiliki extreme distribution

```python
# Membuat num_pipeline dengan QuantileTransformer, SimpleImputer, dan MinMaxScaler
num_pipeline2 = Pipeline(steps=[
    ('outlier', QuantileTransformer(output_distribution='normal')),  # Transformasi dengan QuantileTransformer
    ('imputer', SimpleImputer(strategy='median')),  # Simple Imputer using median strategy
    ('scaler', MinMaxScaler())  # Min-Max Scaling
])

# Mengakses kolom menggunakan nama kolom sebagai indeks dengan X_train.loc
num_tr2 = num_pipeline.fit_transform(X_train[ ['avg_frequency_login_days']])
```

melakukan handling outlier serta scaling terhadap data yang memiliki normal distribution

```python
num_pipeline3 = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', MinMaxScaler())
    ])

num_tr3 = num_pipeline3.fit_transform(X_train[['age','joining_date',
 'last_visit_time',
 'days_since_last_login',]])
num_tr3
```

Melakukan scaling terhadap kolom yang tidak memiliki outlier

```python
full_pipeline = ColumnTransformer([
        ('num1', num_pipeline, ['points_in_wallet', 'avg_time_spent', 'avg_transaction_value']),
        ('num2', num_pipeline2, ['avg_frequency_login_days']),
        ('num3', num_pipeline3, ['age','joining_date', 'last_visit_time','days_since_last_login',]),
        ('cat1', cat_pipeline, ['gender','region_category','membership_category','joined_through_referral', 'preferred_offer_types',
 'medium_of_operation','internet_option','used_special_discount','offer_application_preference','past_complaint','complaint_status','feedback']),
    ])
train_tr = full_pipeline.fit_transform(X_train)
```

# **Modeling**

##Sequential

Pada section ini kita menggunakan model sequential untuk melakukan prediksi. Lalu kita melakukan improvement dengan menentukan dense serta input shape dll. Lalu pada pemodelan ini juga digunakan epocs=10 serta melakukan plot grafik visualisasi untuk hasil dari pemodelan.

```python
# pipeline ETL tensorflow
train_set = tf.data.Dataset.from_tensor_slices((X_train_clean, y_train))

#shuffle train set
train_set = train_set.shuffle(buffer_size=712).batch(64)

#val dataset
val_set = tf.data.Dataset.from_tensor_slices((X_val_clean, y_val))
val_set = val_set.batch(64)
```

```python
alp = 10
inp = X_train_clean.shape[1]
out = 1
hid = int(len(X_train_clean)/(alp*(inp+out)))
print(f'Jumlah neuron pada hidden layer: {hid}')

model_sequential = tf.keras.models.Sequential([
    tf.keras.layers.Dense(inp, activation='relu', input_shape=(inp,)),
    tf.keras.layers.Dense(hid, activation='relu'),
    tf.keras.layers.Dense(out, activation='sigmoid')
])
model_sequential.summary()
```

```python
tf.keras.utils.plot_model(model_sequential, show_shapes=True)
```

```python
model_sequential.compile(loss='binary_crossentropy',optimizer='Adam',metrics='accuracy')
```

```python
history = model_sequential.fit((train_set), epochs=100, validation_data=(val_set))
```

```python
epoch=100

fig, ax = plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,epoch+1),history.history['loss'],label='loss')
ax[0].plot(range(1,epoch+1),history.history['val_loss'],label='val_loss')
ax[0].set_title('Loss')
ax[0].legend()
ax[0].set_xlabel('Epoch')
#ax[0].set_ylim(0,0.01)

ax[1].plot(range(1,epoch+1),history.history['accuracy'],label='accuracy')
ax[1].plot(range(1,epoch+1),history.history['val_accuracy'],label='val_accuracy')
ax[1].set_title('Accuracy')
ax[1].legend()
ax[1].set_xlabel('Epoch')
#ax[1].set_ylim(0.5,1.5)

plt.show()
```

Berdasarkan grafik diatas model sudah cukup baik namun masih ada spike walaupun tidak terlalu significant.

## functional

Pada section ini saya melakukan prediksi dengan model functional dan nantinya akan dilakukan model improvement serta memunculkan plot grafik dari pemodelannya.

```python
# functional API
input_layer= tf.keras.Input(shape=(X_train_clean.shape[1:]),name='input_layer')

hidden_1= Dense(30, activation='relu', name='hidden_1')(input_layer)
hidden_2= Dense(30, activation='relu', name='hidden_2')(hidden_1)
hidden_3= Dense(10, activation='relu', name='hidden_3')(hidden_2)
concat= tf.keras.layers.concatenate([input_layer,hidden_3])

output_layer= Dense(1, activation='sigmoid', name='output_layer')(concat)

model_functional= Model(inputs=input_layer,
                  outputs=output_layer)

model_functional.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
tf.keras.utils.plot_model(model_functional, show_shapes=True)
```

```python
history2 = model_functional.fit((train_set), epochs=100, validation_data=(val_set))
```

```python
epoch2=100

fig, ax = plt.subplots(ncols=2,figsize=(16,5))
ax[0].plot(range(1,epoch2+1),history2.history['loss'],label='loss')
ax[0].plot(range(1,epoch2+1),history2.history['val_loss'],label='val_loss')
ax[0].set_title('Loss')
ax[0].legend()
ax[0].set_xlabel('Epoch')
#ax[0].set_ylim(0,0.01)

ax[1].plot(range(1,epoch2+1),history2.history['accuracy'],label='accuracy')
ax[1].plot(range(1,epoch2+1),history2.history['val_accuracy'],label='val_accuracy')
ax[1].set_title('Accuracy')
ax[1].legend()
ax[1].set_xlabel('Epoch')
#ax[1].set_ylim(0.5,1.5)

plt.show()
```

```python
# functional API improvement
input_layer= tf.keras.Input(shape=(X_train_clean.shape[1:]),name='input_layer')

hidden_1_imp= Dense(60, activation='relu', name='hidden_1_imp')(input_layer)
dropout_layer1 = Dropout(0.3)(hidden_1_imp)
hidden_2_imp= Dense(30, activation='relu', name='hidden_2_imp')(dropout_layer1)
dropout_layer2 = Dropout(0.3)(hidden_2_imp)
hidden_3_imp= Dense(10, activation='relu', name='hidden_3_imp')(dropout_layer2)
concat= tf.keras.layers.concatenate([input_layer,hidden_3_imp])

output_layer= Dense(1, activation='sigmoid', name='output_layer')(concat)

model_func_imp= Model(inputs=input_layer,
                  outputs=output_layer)

model_func_imp.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# **Model Evaluation**

```python
def class_report(Model):
    if (Model == Model):
        print(classification_report((Model.predict(X_test_clean)>0.5).astype(int).tolist(),y_test))
    else:
        print(classification_report(Model.predict(X_test_clean),y_test))
```

```python
def cmatrix(Model,title):
    if (Model==Model):
        matrix1 = confusion_matrix((Model.predict(X_test_clean)>0.5).astype(int).tolist(),y_test)
        sns.heatmap(matrix1, cmap = 'Blues',annot = True, fmt = 'd')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title(title)
        plt.show
    else:
        matrix2 = confusion_matrix(Model.predict(X_test_clean),y_test)
        sns.heatmap(matrix2,cmap = 'Blues',annot = True , fmt = 'd')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.title(title)
        plt.show
```

Fungsi2 diatas digunakan untuk membantu kita melakukan evaluasi dan hanya tinggal memanggil fungsinya.


*   Fungsi pertama dibuat untuk melihat class report dari model agar kita dapat melihat hasil evaluasi dari setiap model yang telah kita buat
*   Fungsi kedua digunakan untuk membuat conf matriks dari model yang telah kita buat.

Dan metriks evaluasi yang digunakan adalah akurasi dikarenakan data yang kita miliki balance sehingga kita bisa menggunakan metrics evaluasi accuracy.
