```python
import pandas as pd

cb_df=pd.read_csv('CB.csv',delimiter=';')
lb_df=pd.read_csv('LB.csv',delimiter=';')
rb_df=pd.read_csv('RB.csv',delimiter=';')
lw_df=pd.read_csv('LW.csv',delimiter=';')
rw_df=pd.read_csv('RW.csv',delimiter=';')
p_df=pd.read_csv('P.csv',delimiter=';')
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns

dataframes = [cb_df, lb_df, rb_df, lw_df, rw_df, p_df]

for df_name, df in zip(['cb_df', 'lb_df', 'rb_df', 'lw_df', 'rw_df', 'p_df'], dataframes):
    print(f"--- Pearson Correlation for {df_name} ---")

    target = df['GOAL']

    corr_values = []
    for col in df.columns:
        if col != 'GOAL':
            corr, _ = pearsonr(df[col], target)
            corr_values.append((col, corr))

    corr_values.sort(key=lambda x: -x[1])

    print(corr_values)

    plt.figure(figsize=(10, 8))
    correlation_matrix = df.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt=".2f")
    plt.title(f"Pearson Correlation Heatmap for {df_name}")
    plt.show()
```
