```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  
```

```python
df = pd.read_csv('validation_resultsacc.csv', delimiter=',')
df2 = pd.read_csv('validation_resultsshot.csv', delimiter=',')
```

```python
kolom_r2 = 'R2_Test'
kolom_mse = 'MSE'
kolom_mae = 'MAE'


# Create a figure with subplots
fig, axes = plt.subplots(ncols=3, figsize=(12, 6))  # 3 columns of subplots

# Create the IQR-based box plot for R2 using seaborn
sns.boxplot(y=df[kolom_r2], ax=axes[0])
axes[0].set_title(f'LBACC R-Squared')
axes[0].set_xlabel('')
axes[0].set_ylabel('R-Squared')

# Create the IQR-based box plot for MSE using seaborn
sns.boxplot(y=df[kolom_mse], ax=axes[1])
axes[1].set_title(f'LBACC {kolom_mse}')
axes[1].set_xlabel('')
axes[1].set_ylabel(kolom_mse)

# Create the IQR-based box plot for MAE using seaborn
sns.boxplot(y=df[kolom_mae], ax=axes[2])
axes[2].set_title(f'LBACC {kolom_mae}')
axes[2].set_xlabel('')
axes[2].set_ylabel(kolom_mae)

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
